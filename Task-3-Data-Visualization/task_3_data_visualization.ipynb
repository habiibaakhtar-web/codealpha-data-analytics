```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('books_full_dataset.csv')
```

```python
df.head()
```

```python
df["Rating"].value_counts().plot(kind="bar", figsize=(8,5))

plt.title("Distribution of Book Ratings")
plt.xlabel("Rating")
plt.ylabel("Number of Books")
plt.show()
```

### Insight from Rating Distribution Chart
The bar chart shows that One-star books are the most common category in the dataset, while Two-star and Three-star books are the least frequent. This indicates that the dataset is slightly skewed toward lower-rated books rather than being perfectly balanced across all rating categories.

```python
df["Price"] = df["Price"].str.replace("Â£", "", regex=False)
df["Price"] = df["Price"].astype(float)

df["Price"].head()
```

```python
avg_price = df.groupby("Rating")["Price"].mean()

avg_price.plot(kind="bar", figsize=(8,5))

plt.title("Average Price by Rating")
plt.xlabel("Rating")
plt.ylabel("Average Price (£)")
plt.show()
```

### Insight from Average Price by Rating Chart
The chart shows that average book price does not consistently increase with rating. Three-star books have the highest average price, while Four-star books have the lowest. This suggests that rating is not a strong indicator of price in this dataset.


```python
plt.figure(figsize=(8,5))
plt.hist(df["Price"], bins=6)

plt.title("Distribution of Book Prices")
plt.xlabel("Price (£)")
plt.ylabel("Frequency")
plt.show()
```

### Insight from Price Distribution Histogram
The histogram shows how book prices are distributed across different price ranges. Most books appear to fall within the mid-price range rather than at the extreme low or high ends. This suggests that the dataset is centered around moderately priced books, with fewer books at the cheapest and most expensive ranges.

```python
plt.figure(figsize=(8,5))
plt.boxplot(df["Price"])

plt.title("Boxplot of Book Prices")
plt.ylabel("Price (£)")
plt.show()
```

### Insight from Price Boxplot
The boxplot provides a compact view of the spread of book prices and helps identify potential outliers. It shows the median price, the central range of prices, and whether any books fall unusually far from the rest of the dataset. This makes it useful for detecting anomalies and understanding price variability.
