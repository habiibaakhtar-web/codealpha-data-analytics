```python
import requests
```

```python
from bs4 import BeautifulSoup
```

```python
import pandas as pd
```

```python
url = "http://books.toscrape.com"
```

```python
response = requests.get(url)
```

```python
print(response.status_code)
```

```python
html_content = response.text
```

```python
print(html_content[:1000])
```

```python
soup = BeautifulSoup(response.text , "html.parser")
```

```python
print(type(soup))
```

```python
book = soup.find("h3")
print(book)
```

```python
book_title = book.a["title"]
print(book_title)
```

```python
books = soup.find_all("h3")

for book in books:
    title = book.a["title"]
    print(title)
```

```python
books = soup.find_all("h3")

book_titles = []

for book in books:
    title = book.a["title"]
    book_titles.append(title)

print(book_titles)
```

```python
import pandas as pd

df = pd.DataFrame(book_titles, columns=["Book Title"])

df.head()
```

```python
df.to_csv("books_dataset.csv", index=False)
```

```python
books = soup.find_all("article", class_="product_pod")

data = []

for book in books:
    title = book.h3.a["title"]
    price = book.find("p", class_="price_color").text
    rating = book.p["class"][1]
    availability = book.find("p", class_="instock availability").text.strip()

    data.append([title, price, rating, availability])

data
```

```python
import pandas as pd

df = pd.DataFrame(data, columns=["Title", "Price", "Rating", "Availability"])

df.head()
```

```python
df.to_csv("books_full_dataset.csv", index=False)
```
