# Task 02 - Exploratory Data Analysis (EDA)

## Questions Asked
1. Which rating is most common?
2. What is the average price of books?
3. What are the minimum and maximum book prices?
4. Does book rating affect price?
5. Are there any data quality issues?

## Data Structure
- Rows: 20
- Columns: 4
- Variables: Title, Price, Rating, Availability

## Data Quality Checks
- No missing values found.
- Price column contained encoding issues (Â£).
- Price column was converted from text to numeric format.

## Key Findings
- Average book price: £38.05
- Minimum price: £13.99
- Maximum price: £57.25
- One-star books are the most common rating.
- Two-star and Three-star books are the least common ratings.

## Hypothesis Testing
Hypothesis:
Higher-rated books cost more.

Result:
The hypothesis was rejected because there is no consistent increase in price with higher ratings.

## Conclusion
The dataset was successfully cleaned and analyzed. No missing values were found, but the Price column required preprocessing. Rating does not appear to have a strong relationship with price in this dataset.
